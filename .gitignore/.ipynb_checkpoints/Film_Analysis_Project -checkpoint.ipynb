# Cleaning the Data:

## Importing the libraries and various data files

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# these are the various libraries we will need
```

```python
movie_budgets_df = pd.read_csv('zippedData/tn.movie_budgets.csv.gz')     
title_basics = pd.read_csv('zippedData/imdb.title.basics.csv.gz')   
title_ratings = pd.read_csv('zippedData/imdb.title.ratings.csv.gz')  
tmdb = pd.read_csv('zippedData/tmdb.movies.csv.gz')
# here is the raw data I will be using
```

## Assigning genres to the genres column

```python
# firstly, the tmbd database has it's genre's encoded so we must decode the 
# genre's using the below dictionary

genres={'28':"Action",
'12':"Adventure",
'16':"Animation",
'35':"Comedy",
'80':"Crime",
'99':"Documentary",
'18':"Drama",
'10751':"Family",
'14':"Fantasy",
'36':"History",
'27':"Horror",
'10402':"Music",
'9648':"Mystery",
'10749':"Romance",
'878':"Science Fiction",
'10770':"TV Movie",
'53':"Thriller",
'10752':"War",
'37':"Western"}

tmdb['genre_ids'] = tmdb['genre_ids'].replace(pd.Series(genres).astype(str), 
                                              regex=True)
tmdb['genre_ids']=tmdb['genre_ids'].str.strip('[]').str.split(', ')

# creating a dictionary and using the key, values pairs to decode the genres
```

## Making merges to bring the data together. 

```python
# secondly, we must merge the two IMDB dataframes

title_master = title_basics.merge(title_ratings, how='inner')
title_master.head()
```

```python
# making the title and tconst ID part of the same object so I can drop 
# douplicates effectively

title_master['title_and_id'] = title_master['primary_title'] +' id ' + \
title_master['tconst']
```

```python
# we must merge the tmdb database with the imdb database. 
# we chose to do an inner merge, here because we 
# are only interested in the movies that these dataframes have in common

tmdb_and_imdb = tmdb.merge(title_master, how = 'inner')
tmdb_and_imdb.head()
```

```python
# the next dataframe we want to fold in has the column "movie" but we want 
# to rename it to title so the merge can happen on the title of the film, 
# so we must rename the column

movie_budgets_df = movie_budgets_df.rename(columns={'movie':'title'})
```

```python
# now we want to merge the fincial dataframe from The Numbers 
# with all the others based on Title, so we specify that 
# with the on = 'title'

master_unclean = tmdb_and_imdb.merge\
(movie_budgets_df, on = 'title', how = 'inner')
master_unclean.head()
```

```python
# let's take a look at the amount of information we have now and the types 
# of information in our columns we can see where we have duplicate columns, 
# we now have x and y versions to distinguish them from each other
# runtime_minutes and genres are the only columns that do not have 
# 16669 non null values after this merge

master_unclean.info()
```

## Limiting the data on runtime 

```python
# now we want to exclude movies from this master dataframe which are less 
# than 45 min and more than three hours, because we want to base the 
# recomendations on films that are typically seen in the movie theater
# and put that information on a new dataframe called run_time

run_time = master_unclean.loc[(master_unclean['runtime_minutes'] > 45) \
                              & (master_unclean['runtime_minutes'] < 180)]
run_time.head()
```

## Dropping unnecessary columns 

```python
# here, we want to drop some columns where the information is irrelivant 
# or duplicated. I'm dripping id_x and id_y since I don't need another unqie 
# identifyer,  title, original title since I already have the title column I 
# need,  start year, year and release_date_x since I already have a datetime 
# column, genres since the genre_ids column I already have is more robust, 
# and all the rating a popularity information since I won't be using that data 
# for my analysis

df = master_unclean.drop\
(['Unnamed: 0', 'id_x', 'original_title', 'start_year', 'id_y',\
'genres', 'release_date_x', 'primary_title', 'original_language', \
'popularity', 'vote_average', 'vote_count', 'numvotes', 'averagerating',
  'title', 'tconst'],
 axis = 1)
```

```python
#let's take a look at where our dataframe is right now

df.head()
```

## Formatting the Remaining Data 

```python
# this function will take a string value with commas and dollar signs and 
# return an integer value in a format that I can do math upon 

def reformat(string_to_money):
    string_to_money = string_to_money.replace(',', '').replace('$', '')
    string_to_money = int(string_to_money)
    return string_to_money
```

```python
#reformatting the production budget, domestic gross and worldwide gross columns 

df['production_budget'] = df['production_budget'].map(reformat)
df['domestic_gross'] = df['domestic_gross'].map(reformat)
df['worldwide_gross'] =df['worldwide_gross'].map(reformat)
```

```python
#checking to make sure my finacial data is in integer form

df.info()
```

```python
# creating total gross, net_profit and ROI as finacial metrics. I will be using 
# net profit as my main metric but I wanted to create total gross and ROI 
# for any additional analysis in my conclusion 

df['total_gross'] = df['domestic_gross'] + df['worldwide_gross']
df['net_profit']= df['total_gross'] - df['production_budget'] 
df['ROI'] = (df['net_profit'] / df['production_budget']) * 100
```

## Checking for Null Values

```python
# seeing how many columns have null values

df.isna().sum()
```

I'm going to drop the rows where the 
runtime data is missing, it's a small section of my dataset and since I will be making recommendations based on runtime, I don't want to throw the data off.

```python
# dropping rows with null values 

df.dropna(axis=0, inplace= True) 
```

```python
#to check if the 34 rows dropped and we have no null values

df.isna().sum()
```

```python
df.info()
```

## Removing duplicates

```python
# to show how many duplicates we have as a slice of the current df

df[df.duplicated(subset = df.drop('genre_ids', axis = 1).columns, 
                )].sort_values(by = 'title_and_id')
```

```python
# dropping duplicate rows that are about the same film by using the above 
# subset (which checks for everything except matching values in the 
# 'genre_ids' columns.) 

df.drop_duplicates(subset = df.drop('genre_ids', axis = 1).columns,
                   inplace = True) 
```

```python
df.info()
```

```python
df
```

After cleaning and removing null values and duplicates, this analysis will use a dataset the size of 2,864 films. 

# Recommendations:

## Genre recommendations 

```python
#first we need to use .explode to make a row for each genre that each film
#falls into and so we can compare genres side by side based on total gross, 
#net profit and ROI 

df_exploded = df.explode('genre_ids', ignore_index=False)
df_exploded.head()
```

```python
df_exploded.info()
```

```python
null_genre = df_exploded.loc[df_exploded['genre_ids'] == '']
null_genre
```

```python
df_exploded.drop(labels = null_genre.index, inplace = True) 
```

```python
df_exploded.info()
```

```python
# making a table that shows all the genres and their adverage values,
# sorted by net profit

by_genre_net = df_exploded.groupby(by='genre_ids', axis=0).mean()\
.sort_values('net_profit', ascending=False)
by_genre_net
```

```python
# making a plot to visualize the above table so we can easily see which 
# genres dominate net profits 

values = by_genre_net['net_profit'] 
colors = ['grey' if (x < 2.838668e+08	) else 'blue' for x in values ]

sns.set_style(style = 'white')

plt.figure(figsize=(8,8))

ax = sns.barplot(data = df_exploded, x = "net_profit", y = "genre_ids", 
           order = by_genre_net.index, palette=colors)
ax.set(xlabel="Average Net Profit in Hundred Millions", ylabel = "Genre", 
       title = "Net Profit by Genre")

# making sure my labels don't get cut-off

plt.tight_layout();

# specifying the scientific notation to units in regular English 
# investigate this blank bar
```

The takeaway here is that animation, adventure, fantasy and family are the top four genres by close to 70 million dollars.

## Release Date 

```python
# making a new series called release_date that only contains films if they are 
# in the Animation, Adventure, Fantasy or Family genres for further analysis 

release_date = df_exploded['genre_ids'].isin(['Animation', 'Adventure',
                                              'Fantasy', 'Family'])

release_date.value_counts()
```

```python
# slicing out the series from df_exploded and making it it's own dataframe

release_date = df_exploded[release_date]
release_date.head()
```

```python
# turning release_date_y into datetime format 

release_date['date_time'] = pd.to_datetime(release_date['release_date_y'])
release_date.info()
```

```python
#isolating the release month from that datetime column 

release_date['month_num'] = release_date['date_time'].dt.month
release_date.head()
```

```python
# putting a column for the name of the month for readability 

release_date['month_name'] = release_date['date_time'].dt.month_name()
release_date.head()
```

```python
#making a pivot table to aggrigate the data based on genre

pivot = pd.pivot_table(release_date, values='net_profit', index='month_name', 
                       columns='genre_ids')

pivot
```

```python
pivot['total'] = pivot['Adventure'] + pivot['Animation'] + \
                     pivot['Family'] +  pivot['Fantasy']
pivot
```

```python
pivot.sort_values('total', inplace = True, ascending=False)
pivot
```

```python
pivot.drop(columns = 'total', inplace = True)
pivot
```

```python
# making a plot to show the above pivot table as a stacked bar chart

ax = pivot.plot(kind='bar', figsize=(7, 7), stacked=True)

#ax.axhline(1500000000, color = "blue")
#ax.axhline(1900000000, color = "green")

ax.set(xlabel = "Month", ylabel = "Net Profit in Billions", 
       title = "Average Net Profit By Genre By Month")

plt.xticks(rotation = 45);
```

The takeaway here is to release films in June, where the average film in this genre made over 1.9 billion dollars (shown by the green line.) If there is a delay in production due to Covid-19, I'd recommend these months as backup choices: May, March, July, November or April. 

Interesting find: Fantasy films do not do well in July compared to other high earning months. 

## Runtime of Film 

```python
release_date.info()
```

```python
# all movies in this data set are between 30 and 180

plt.figure(figsize=(10,10))

runtime_plot = sns.regplot(data = release_date, x = 'runtime_minutes', 
                 y = 'net_profit', color = 'green',
                 scatter_kws={'alpha':0.3})

ax.set(xlabel = 'Runtime Minutes', ylabel = 'Net Profit in Billions', 
      title = "Net Profit of Movies by Runtime Minutes");
```

This doesn't tell us very much so instead let's sort runtime minutes into categorical data instead

```python
bins = [ 60, 90, 120, 150, 180]
bins
```

```python
labels = ['Sixty', 'One hundred twenty',
          'One hundred fifty', 'One hundred eighty']
```

```python
cut_data, bin_edges = pd.cut(release_date['runtime_minutes'], bins = bins,
                              labels = labels, retbins = True)
bin_edges

release_date['runtime_bins']= cut_data
release_date
```

```python
runtime_df = release_date.groupby(by='runtime_bins').mean()
runtime_df = runtime_df.reset_index()
runtime_df
```

```python
runtime_df['net_profit']
```

```python
runtime_df.sort_values(by = 'net_profit', inplace = True) 
runtime_df
```

```python
plt.figure(figsize=(8,8))

ax = sns.barplot(data = release_date, x = 'runtime_bins', y = 'net_profit')
plt.xticks(rotation=45)

ax.set(xlabel = 'Runtime Bins', ylabel = 'Average Net Profit in 100K Dollars', 
       title = "Net Profit by Runtime Bin")
    
plt.tight_layout();
# this will help my labels not get cut-off 
```

Here the important piece of information is that films between 120-150 minutes do better consistently, and movies between 150-180 movies do about as well on average but have a much higher rate of deviation. (It's more likely that films in that range will do either much better or much worse than films in the previous bin. 

# Conclusion: 

## Three Business Recommendations: 

My three business recommendations to Microsoft Studios would be: 
1) Focus on making films that are a combination of animation, fantasy, adventure, and family. There are many examples of the combination of these genres already, (Pixar Studios being the best example of having luck here.) 

2) Release this film in June, May, March of any genre, or July with an emphasis on non-Fantasy films (which, interestingly do much better in November.)

3) Make the movie somewhere within 120-150 minutes long since movies in this range tend to make the most net profit on average with the most consistency.
